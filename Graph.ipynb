```python
import numpy as np
import matplotlib.pyplot as plt

# Hypothetical stock prices for Netflix (for demonstration purposes)
dates = np.arange(1, 31)  # Assume days of a month
actual_prices = np.array([450, 455, 460, 465, 470, 475, 473, 478, 480, 485, 482, 488, 490, 495, 492, 498, 500, 505, 503, 508, 510, 515, 512, 517, 520, 525, 523, 530, 535, 540])
predicted_prices = np.array([448, 453, 459, 463, 468, 474, 472, 476, 479, 484, 481, 486, 489, 494, 490, 496, 499, 504, 501, 506, 509, 513, 510, 515, 518, 523, 521, 528, 533, 538])

# Create a plot
plt.figure(figsize=(10, 6))
plt.plot(dates, actual_prices, label='Actual Prices', marker='o', color='blue')
plt.plot(dates, predicted_prices, label='Predicted Prices', linestyle='--', marker='x', color='red')

# Adding labels and title
plt.title('Netflix Stock Price Prediction')
plt.xlabel('Days')
plt.ylabel('Stock Price (USD)')
plt.legend()

# Show the graph
plt.grid(True)
plt.tight_layout()
plt.show()
```
